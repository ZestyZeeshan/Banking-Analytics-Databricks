# Banking Analytics Dashboard

Objective:
Create business dashboards using Gold Layer tables.

KPIs:
1. Customer Analysis
2. Branch Analysis
3. Loan Analysis
4. Transaction Analysis
5. Fraud Monitoring

```python
#total deposit KPIs
customer_summary = spark.table(
    "banking_catalog.gold.customer_summary"
)

display(customer_summary)
```

```python
#Now lets strt with creating KPI
from pyspark.sql.functions import sum

customer_summary.agg(
    sum("total_balance")
).show()
```

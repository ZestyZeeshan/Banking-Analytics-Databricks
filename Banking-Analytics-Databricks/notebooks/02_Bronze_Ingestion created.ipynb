```python
customers_df = spark.read.csv(
    "/Volumes/banking_catalog/bronze/raw_files/customers.csv.txt",
    header=True,
    inferSchema=True
)

```

```python
display(customers_df)

```

```python
customers_df.printSchema()

```

```python
customers_df.count()
```

```python
customers_df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("banking_catalog.bronze.customers")
```
